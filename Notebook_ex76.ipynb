- Using train_test_split, split the data set in a train set and test set (20%) with `shuffle=True` and `random_state=43`.


````python
iris = load_iris()

X = pd.DataFrame(data=iris['data'], columns=iris.feature_names)
y = pd.DataFrame(data=iris['target'], columns=['target'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2 ,shuffle=True, random_state=43)
y
````

1. Create a function that takes as input the data and returns three **trained** classifiers.

- `clf0` takes as input a binary data set where the class 1 is `0` and class 0 is `1` and `2`.
- `clf1` takes as input a binary data set where the class 1 is `1` and class 0 is `0` and `2`.
- `clf2` takes as input a binary data set where the class 1 is `2` and class 0 is `0` and `1`.

````python
def train(X_train, y_train):
    clf0 = LogisticRegression()
    clf1 = LogisticRegression()
    clf2 = LogisticRegression()
    
    y0 = (y_train['target'] == 0).astype(int) 
    y1 = (y_train['target'] == 1).astype(int)
    y2 = (y_train['target'] == 2).astype(int)
    
    clf0.fit(X_train, y0)
    clf1.fit(X_train, y1)
    clf2.fit(X_train, y2)
    
    return clf0, clf1, clf2
````

````python
clf0, clf1, clf2 = train(X_train, y_train)
````

Compare the results with Logistic Regression algorithm from scikit-learn used in One-vs-All mode. The results may change because the solver may not converge. - `clf1` outputs the probability to belong to the class 1 which is `1`.
- `clf2` outputs the probability to belong to the class 1 which is `2`.

The predicted class is the one that gets the **highest probability** among the three models.

````python
def predict_one_vs_all(X, clf0, clf1, clf2):
    prob_0 = clf0.predict_proba(X)[:, 1]
    prob_1 = clf1.predict_proba(X)[:, 1]
    prob_2 = clf2.predict_proba(X)[:, 1]
    
    all_probs = np.column_stack((prob_0, prob_1, prob_2))
    
    classes = np.argmax(all_probs, axis=1)
    
    return classes
````

````python
clf_sklearn = OneVsRestClassifier(LogisticRegression())
clf_sklearn.fit(X_train, y_train)
y_pred_sklearn = clf_sklearn.predict(X_test)
y_pred_custom = predict_one_vs_all(X_test, clf0, clf1, clf2)

print(f"Custom OvR Accuracy:  {accuracy_score(y_test, y_pred_custom):.2%}")
print(f"Sklearn OvR Accuracy: {accuracy_score(y_test, y_pred_sklearn):.2%}")

identical_predictions = np.array_equal(y_pred_custom, y_pred_sklearn)
print(f"Are the predictions exactly the same? {identical_predictions}")
````
