```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

1. On the same plot, plot the sigmoid function and the custom sigmoids defined as:

- `sigmoid1(x) = 1/(1+ exp(-(0.5*x + 3)))`

- `sigmoid2(x) = 1/(1+ exp(-(5*x + 11)))`

- Add a line representing the probability=0.5

The plot should look like this:

![ex2q1](./w2_day2_ex2_q1.png)

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid1(x):
    return 1 / (1 + np.exp(-(0.5 * x + 3)))

def sigmoid2(x):
    return 1 / (1 + np.exp(-(5 * x + 11)))

x = np.linspace(-20, 15, 5)

y_std = sigmoid(x)
y1 = sigmoid1(x)
y2 = sigmoid2(x)

# Create the plot
plt.figure(figsize=(10, 6), dpi=150)
plt.plot(x, y_std, label='Standard Sigmoid', color='blue')
plt.plot(x, y1, label='sigmoid1(x) (0.5x + 3)', color='orange')
plt.plot(x, y2, label='sigmoid2(x) (5x + 11)', color='green')

# Add the 0.5 probability line
plt.axhline(y=0.5, color='red', alpha=0.5, label='Probability = 0.5')

# Formatting
plt.title('Sigmoids')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
